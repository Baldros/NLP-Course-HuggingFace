# Viés e limitações
  
    embora esses modelos pré-treinados sejam ferramentas poderosas, eles têm
    limitações. A maior delas é que, para possibilitar o pré-treinamento em
    grandes volumes de dados, os pesquisadores muitas vezes coletam todo o
    conteúdo que podem encontrar, incluindo o melhor e o pior disponível na
    internet. Falamos sobre esse tipo de problema na parte da apresentação
    de modelos pré-treinados.

    Para dar uma ilustração rápida, vamos voltar ao exemplo de pipeline de
    máscara de preenchimento com o modelo BERT:

```python
# Importando a API utilizada:
from transformers import pipeline
```

```python
# Instanciando modelo:
unmasker = pipeline("fill-mask", model="bert-base-uncased")
```

```python
# Gerando resultados:
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])
```

```python
result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

    Ao ser solicitado para preencher a palavra faltante nessas duas frases,
    o modelo fornece apenas uma resposta livre de gênero (garçom/garçonete).
    As outras são ocupações geralmente associadas a um gênero específico - e
    sim, "prostituta" acabou entre as 5 principais possibilidades que o modelo
    associa com "mulher" e "trabalho". Isso acontece mesmo que o BERT seja um
    dos raros modelos Transformer não construídos coletando dados de toda a
    internet, mas sim usando dados aparentemente neutros (é treinado nos
    conjuntos de dados da Wikipédia em inglês e do BookCorpus).

    Ao utilizar essas ferramentas, é importante ter em mente que o modelo
    original que você está utilizando pode facilmente gerar conteúdo sexista,
    racista ou homofóbico. E mesmo ajustar finamente o modelo com seus
    próprios dados não fará com que esse viés intrínseco desapareça.
